```python
# b. Understanding how many new customers acquired every month 

# Merge orders with customers to get customer_unique_id
cust_orders = pd.merge(left=orders,right=customers[['customer_id', 'customer_unique_id']],on='customer_id',how='inner')

# Extract order month
cust_orders['order_month'] = cust_orders['order_purchase_timestamp'].dt.to_period('M')

# Step 1: Find the first order month for each customer_unique_id
first_order_month = cust_orders.groupby('customer_unique_id')['order_month'].min().reset_index()
first_order_month.rename(columns={'order_month': 'first_purchase_month'}, inplace=True)

# Step 2: Merge back to original data
cust_orders = pd.merge(cust_orders, first_order_month, on='customer_unique_id', how='left')  # here we use customer_unique_id because it remains Same for the same real-world person, even if they place multiple orders or their customer_id changes.

# Step 3: Filter customers whose first purchase is in that month, if yes then they are new customers otherwise existing 
new_customers = cust_orders[cust_orders['order_month'] == cust_orders['first_purchase_month']]    # someone who place their first order in a month called new customer

# Step 4: Count unique new customers per month
new_customers_count = new_customers.groupby('order_month')['customer_unique_id'].nunique().reset_index()    # the key present in groupby gives always result default in ascending order
new_customers_count.columns = ['Month', 'New_Customers']
display(new_customers_count.T)


new_customers_count['Month'] = new_customers_count['Month'].astype(str)   # initially i got error so i use it because Month column is a pandas Period object (Period[M]) and Seaborn/Matplotlib can’t directly plot Period

plt.figure(figsize=(12,6))
sns.barplot(data=new_customers_count, x='Month', y='New_Customers', color='lightgray')
sns.lineplot(data=new_customers_count, x='Month', y='New_Customers', marker='o', color='red')
plt.title("New Customers Acquired per Month", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Number of New Customers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# c. Understand the retention of customers on month on month basis  (which customers come back after their first purchase on month on month basis)

cust_orders = pd.merge(orders[['order_id', 'customer_id', 'order_purchase_timestamp']],customers[['customer_id', 'customer_unique_id']],
                       on='customer_id',how='inner')

cust_orders['order_month']=cust_orders['order_purchase_timestamp'].dt.to_period('M')

first_purchase= cust_orders.groupby('customer_unique_id')['order_month'].min().reset_index()   # custome_unique_id will alwahys remain same for multiple different orders of customer  

first_purchase.rename(columns={'order_month':'cohort_month'},inplace=True)   # cohort_month mean first_purchase month ( customer's first interacted with business)

cust_orders = pd.merge(cust_orders,first_purchase,on='customer_unique_id',how='left')  # here we can use both left or inner join because first_purchase is derrived from cust_orders so every customer_unique_id of cust_order is present in first_purchase  

retention_cust_count = cust_orders.groupby(['cohort_month', 'order_month'])['customer_unique_id'].nunique().unstack().fillna(0)  # after purchasing first time how many times customer return again, here unstack pivots the second columns (order_month) into index instead of rows and this gives the number of unique customers from that cohort who purchased in each month.
display(retention_cust_count)

# visualize it 
mask = np.tri(retention_cust_count.shape[0], retention_cust_count.shape[1], k=0, dtype=bool)   # mask is used to hide certain parts of the matrix when plotting seaborn heatmap ( k= diagonal offset

plt.figure(figsize=(14,6))
sns.heatmap(retention_cust_count, mask = mask, annot=True, fmt=".0f", cmap="Blues", cbar=True)

plt.title("Customer Retention by Cohort on MoM basis",fontsize=16)
plt.xlabel("Order Month")
plt.ylabel("Cohort Month")
plt.xticks(rotation=45)
plt.show()
```

```python
#d. How the revenues from existing/new customers on month on month basis

cust_orders = pd.merge(orders[['order_id', 'customer_id', 'order_purchase_timestamp']],customers[['customer_id', 'customer_unique_id']],
                       on='customer_id',how='inner')

cust_orders['order_month'] = cust_orders['order_purchase_timestamp'].dt.to_period('M')

first_purchase_month = cust_orders.groupby('customer_unique_id')['order_month'].min().reset_index()
first_purchase_month.rename(columns={'order_month':'first_purchase_month'},inplace=True)
first_purchase_month
cust_orders = pd.merge(cust_orders,first_purchase_month,on='customer_unique_id', how = 'left')
cust_orders['cust_type'] = np.where(cust_orders['order_month'] == cust_orders['first_purchase_month'],'new_customers','existing_customers')

order_data = pd.merge(cust_orders,order_items[['order_id','Total_Revenue']], on='order_id',how = 'left')   # to get price 
monthly_revenue = order_data.groupby(['order_month', 'cust_type'])['Total_Revenue'].sum().unstack().fillna(0) 
display(monthly_revenue)

# visualize output
monthly_revenue.index = monthly_revenue.index.astype(str)

fig, axes = plt.subplots(2, 1, figsize=(12,8), sharex=True )

# Plot existing customers
axes[0].plot(monthly_revenue.index, monthly_revenue['existing_customers'], color='blue', marker='o')
axes[0].set_title("Monthly Revenue from Existing Customers")
axes[0].set_ylabel("Revenue")
axes[0].ticklabel_format(style='plain', axis='y')

# Plot new customers
axes[1].plot(monthly_revenue.index, monthly_revenue['new_customers'], color='orange', marker='o')
axes[1].set_title("Monthly Revenue from New Customers")
axes[1].set_ylabel("Revenue")
axes[1].set_xlabel("Order Month")
axes[1].ticklabel_format(style='plain', axis='y')

# Rotate x-axis labels
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Total quantity by category and month

total_category_quantity = pd.pivot_table(orders_items,values='product_id',index='product_category_name',columns='order_month',aggfunc='count',fill_value=0,
                                          margins=True, margins_name='Grand_Total')
print("Total quantity by category :\n")
display(total_category_quantity)

total_category_sales = pd.pivot_table(orders_items,values='price',index='product_category_name',columns='order_month',aggfunc='sum',fill_value=0,margins=True,)
print("\nTotal sales by category :\n")
display(total_category_sales)
```

```python
# Quantity by location
qty_by_location = pd.pivot_table(orders_items,values='product_id',index='customer_state',columns='order_month',aggfunc='count',fill_value=0,margins=True,
                                    margins_name='Total')
print("Total quantity sold by Location:\n")
display(qty_by_location)

sales_by_location = pd.pivot_table(orders_items,values='price',index='customer_state',columns='order_month',aggfunc='sum',fill_value=0,margins=True,margins_name='Total')
print("\nTotal sales by Location:\n")
display(sales_by_location)
```

```python
# Monthly Trends

monthly_trends = orders_items.groupby('order_month').agg( quantity=('product_id','count'),total_sales=('price','sum')).reset_index()
monthly_trends

monthly_trends['order_month'] = monthly_trends['order_month'].astype(str)  # initially i get error that order month dtype was period  it must be string
display(monthly_trends)

# visualize result 
fig, axes = plt.subplots(2, 1, figsize=(12,8), sharex=True )

axes[0].plot(monthly_trends.order_month, monthly_trends['quantity'], color='blue', marker='o')
axes[0].set_title("Monthly Total Quantity Sold")
axes[0].set_ylabel("Quantity_Sold")
axes[0].ticklabel_format(style='plain', axis='y')

axes[1].plot(monthly_trends.order_month, monthly_trends['total_sales'], color='green', marker='o')
axes[1].set_title("Monthly Total Sales")
axes[1].set_ylabel("Salse")
axes[1].set_xlabel("Order Month")
axes[1].ticklabel_format(style='plain', axis='y')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 3. Cross-Selling (Which products are selling together) 
    
Hint: We need to find which of the top 10 combinations of products are selling together in     
each transaction.  (combination of 2 or 3 buying together)
     

##### Market Basket Analysis

```python
# group orders with multiple orders
multi_item_orders = order_items.groupby('order_id').filter(lambda x: len(x)>1)

# group product_id's by order_id and convert them to list go get orders with multiple orders
order_products  = multi_item_orders.groupby('order_id')['product_id'].apply(list)

# generate all possible unique pairs and triplates form order_products
from itertools import combinations    #possible combinations of input iterable (like a list or set), without repetition 
from collections import Counter  #  to show how many times each pair appears appears across all all orders 

# store combinations 
combo_counter_2 = Counter()
combo_counter_3 = Counter()

for product in order_products:
    unique_products = list(set(product))

    # Pairs (2-product combinations)
    for combo in combinations(sorted(unique_products),2):     # combinations(iterable, r)  here r is size of each combinations
        combo_counter_2[combo] +=1
    # triplets (3-products combinations)
    for combo in combinations(sorted(unique_products),3):
        combo_counter_3[combo] +=1 
        
```

```python
top_10_pairs = combo_counter_2.most_common(10)

top_10_triplets = combo_counter_3.most_common(10)

print('top 10 combinations of product_pairs that are selling together with their freq : \n')
display(top_10_pairs)

print('\n top 10 combinations of product_triplets that are selling together with their freq : \n')
display(top_10_triplets)
```

### 5. Customer satisfaction towards category & product 
 

```python
#prepare data for customer_satisfaction...

# Merge review data with orders
review_orders= pd.merge(orders[['order_id', 'customer_id', 'order_month']],review_ratings[['review_id','order_id','review_score','review_creation_date']],
                           on='order_id',how='inner')
# Merge with order items to get product and seller
review_items= pd.merge(review_orders,order_items[['order_id', 'product_id', 'seller_id', 'price']], on='order_id', how = 'inner')
# Merge with products to get category
review_items= pd.merge(review_items,products[['product_id','product_category_name']],on='product_id',how='left')  # use left join because we want to keep all teh review_ratings if product is missing in product table that was in review_items
# Merge with customers to get location
review_items = pd.merge(review_items, customers[['customer_id','customer_zip_code_prefix','customer_state']], on='customer_id', how='left') # use left join to preserve all review rows, even if the customer info (like location/state) is missing in the customers
review_items = pd.merge(review_items, locations[['geolocation_zip_code_prefix','geolocation_city', 'geolocation_state']], left_on = 
                        ['customer_zip_code_prefix','customer_state'], right_on = ['geolocation_zip_code_prefix','geolocation_state'], how='inner')  # here i checked that  some  cusot_state not match with geological_state so we use customer_state also in join_key
#here some product_category_name is not present so safer option is to fillna with unknown 
review_items['product_category_name'] = review_items['product_category_name'].fillna('unknown')
review_items
```

```python
# a. Which categories (top 10) are maximum rated & minimum rated? 

# rating by category
cat_rating = review_items.groupby('product_category_name')['review_score'].agg(['mean','count']).reset_index()
cat_rating.rename(columns={'mean':'avg_rating', 'count':'rating_count'}, inplace=True)

# Global mean rating
C = cat_rating['avg_rating'].mean()

# Choose m (minimum rating count required)      (Small datasets: m = 20–50, Medium datasets: m = 100–500, Large datasets: m = 500–1000
m = 500   # we can try: int(cat_rating['rating_count'].quantile(0.80))  50-80 percentile based of balanced result

# Bayesian average or Weighted Rating formula(IMDb style)
cat_rating['weighted_rating'] = ((cat_rating['rating_count'] / (cat_rating['rating_count'] + m)) * cat_rating['avg_rating']  
                                                      +   (m / (cat_rating['rating_count'] + m)) * C)

# Top & Bottom categories
top_rated_cat = cat_rating.sort_values(by='weighted_rating', ascending=False, ignore_index=True).head(10)
min_rated_cat = cat_rating.sort_values(by='weighted_rating', ascending=True, ignore_index=True).head(10)

print("Top 10 categories with highest ratings (Weighted Rating)\n")
display(top_rated_cat)
print("\n Top 10 categories with lowest ratings (Weighted Rating)\n")
display(min_rated_cat)

#Note : here review count also matters so i used this but it gives fluctuated review_counts with respect to their ratings like 4.3593 with only 103 review
#       count has higher precedence whereas 4.3451 rating with 1075 has lower precedence so use bayesian average or weighted rating (imdb approach ) to solve this
```

```python
# b. Which products (top10) are maximum rated & minimum rated?

product_rating = review_items.groupby('product_id')['review_score'].agg(['mean','count']).reset_index()
product_rating.rename(columns={'mean':'avg_rating', 'count':'rating_count'},inplace=True)

top10_max_products = product_rating.sort_values(by=['avg_rating','rating_count'], ascending=[False,False], ignore_index=True).head(10)
top10_min_products = product_rating.sort_values(by=['avg_rating','rating_count'], ascending=[True,False], ignore_index=True).head(10)

print('Top 10 Products with highest ratings \n')
display(top10_max_products)
print('\n Top 10 Products with lowest ratings \n')
display(top10_min_products)

# visualize result
fig, axes = plt.subplots(1,2, figsize=(16,6))

sns.barplot(data=top10_max_products, x='rating_count',y ='product_id',palette='Greens_r', ax=axes[0])
axes[0].set_title("Top 10 Highest Rated Products (5)")
axes[0].set_xlabel("Rating_Count")
axes[0].set_ylabel("Product_id")

sns.barplot(data=top10_min_products, x='rating_count',y ='product_id',palette='Reds_r', ax=axes[1])
axes[1].set_title("Top 10 Lowest Rated Products (1)")
axes[1].set_xlabel("Rating_Count")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```
